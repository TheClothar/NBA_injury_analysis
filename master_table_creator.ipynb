```python
import pandas as pd
import glob
import os
```

```python
csv_pattern = os.path.expanduser("~/Desktop/NBA_injuries_project/player_stats_season/nba_*_totals.csv")
files = glob.glob(csv_pattern)
print(f" Found {len(files)} files")
```

```python
all_stat = []

for file_path in files:
    #read in the csv
    df = pd.read_csv(file_path)

    #parse season from filename
    basename = os.path.basename(file_path)
    year_str = basename.split("_")[1:3]
    season = f"{year_str[0]}-{year_str[1]}"


     # Add season column
    df["season"] = season


    # Append to list
    all_stat.append(df)

    print("appended to all_stat")
```

```python

# Step 3: Combine all into one DataFrame
all_seasons_stats = pd.concat(all_stat, ignore_index=True)
```

```python
import cloudscraper
import os
import time

scraper = cloudscraper.create_scraper()

years = range(2000, 2025)
url_start = "https://www.basketball-reference.com/leagues/NBA_{}_totals.html"
output_dir = "Desktop/NBA_injuries_project/player_stats"

os.makedirs(output_dir, exist_ok=True)

for year in years:
    url = url_start.format(year)
    print(f"Fetching {url}...")
    
    try:
        response = scraper.get(url)
        
        if response.status_code == 200:
            with open(f"{output_dir}/{year}.html", "w+", encoding='utf-8') as f:
                f.write(response.text)
            print(f"Saved {year}.html ✅")
        else:
            print(f"❌ Failed to fetch {year} (status {response.status_code})")
    
    except Exception as e:
        print(f"⚠️ Error fetching {year}: {e}")
    
    time.sleep(3)
```
